# Carry: all-correct order 

```python
ordered_list = epoch_order_count(all_run_info_add)['{}_carry_epoch'.format(all_correct_case)].most_common()
x_count = 0
for _, x in ordered_list:
    x_count = x_count + x
    
y_list = list()
x_list = list()
for y, x in ordered_list:
    x_list.append(x / x_count * 100)
    y_list.append(y)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
y_pos = np.arange(len(y_list))

ax.barh(y_pos, x_list)
ax.set_yticks(y_pos)
ax.set_yticklabels(y_list)
ax.invert_yaxis()  # labels read top-to-bottom
ax.grid(axis='x')
ax.set_ylabel('Conquest order')
ax.set_xlabel('Percentage of cases (%)')
ax.set_title('The conquest order of carry datasets: addition')

save_fig('conquest_order_of_carry_datasets_add')
```
